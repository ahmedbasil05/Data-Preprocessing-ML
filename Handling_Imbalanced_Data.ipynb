**Imbalanced Data**

a dataset with an unequal class distribution

```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#loading the data
credit = pd.read_csv('/content/credit_data.csv')
```

```python
credit.head()
```

```python
credit.shape
```

```python
#distribution of two classes
credit.Class.value_counts()
```

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
#seperating the legit and fraud transcations
legit = credit[credit['Class'] == 0]
fraud = credit[credit['Class'] == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

**Under Sampling** - handles imbalanced data

samples dataset containig similar distribution of legit and fraud transactions

```python
legit_sample = legit.sample(n=492)
legit_sample.shape
```

**Concatenation** - concatenating two dataframes

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
new_data.shape
```
